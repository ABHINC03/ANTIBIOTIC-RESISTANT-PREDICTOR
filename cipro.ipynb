```python
import pandas as pd
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.model_selection import train_test_split as tts
from sklearn.metrics import accuracy_score,f1_score,roc_auc_score,precision_recall_curve
```

```python
df=pd.read_csv(r'dataset/dataset_ciprofloxacin_final.csv')
```

```python
df
```

```python
X=df.iloc[:,1:]
```

```python
y=df.iloc[:,0]
```

```python
X_train,X_test,y_train,y_test=tts(X,y,test_size=0.2,random_state=42)
```

```python
from sklearn.model_selection import RandomizedSearchCV
rf=ExtraTreesClassifier(random_state=42)
rf_params = {
    "n_estimators": [50, 100, 200, 300, 400, 500],
    "max_depth": [None, 5,10,20, 40,60],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4,10],
    "max_features": ["sqrt", "log2"],
    "bootstrap": [True]
}
search = RandomizedSearchCV(
    rf,
    rf_params,
    n_iter=20,
    refit="accuracy",
    n_jobs=-1,
    cv=5,
    scoring=['accuracy', 'f1', 'roc_auc','recall'],
)

a=search.fit(X_train, y_train)
a
```

```python
et=ExtraTreesClassifier(bootstrap=True,max_depth=60,random_state=42,class_weight='balanced')
```

```python
et.fit(X_train,y_train)
y_pred=et.predict(X_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
f1_score(y_pred,y_test)
```

```python
roc_auc_score(y_pred,y_test)
```

```python
precision_recall_curve(y_pred,y_test)
```
