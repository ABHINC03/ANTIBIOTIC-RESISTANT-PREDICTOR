```python
from Bio.Blast import NCBIWWW
from Bio import SeqIO

# 1. Read your FASTA file
record = SeqIO.read("gene.fasta", format="fasta-blast")

# 2. Run BLAST online (e.g., blastn against the 'nt' database)
print("Running BLAST online. This may take a few minutes...")
result_handle = NCBIWWW.qblast("blastn", "nt", record.seq)

# 3. Save the results to an XML file
with open("my_blast_results.xml", "w") as out_handle:
    out_handle.write(result_handle.read())

print("BLAST complete and saved!")
result_handle.close()
```

```python
from Bio.Blast import NCBIXML

# Open the XML result file we just saved
with open("my_blast_results.xml") as result_handle:
    blast_records = NCBIXML.parse(result_handle)
    
    # Loop through the records (usually one per query sequence)
    for blast_record in blast_records:
        for alignment in blast_record.alignments:
            for hsp in alignment.hsps: # HSP = High-scoring Segment Pair
                # Only show significant matches (E-value < 0.05)
                if hsp.expect < 0.05:
                    print("**** ALIGNMENT ****")
                    print(f"Sequence: {alignment.title}")
                    print(f"Length: {alignment.length}")
                    print(f"E-value: {hsp.expect}")
                    print(f"Score: {hsp.score}")
                    # Print the actual alignment
                    print(hsp.query[0:75] + "...")
                    print(hsp.match[0:75] + "...")
                    print(hsp.sbjct[0:75] + "...\n")
```

```python
from Bio.Blast import NCBIXML

# Your list of 92 model features
model_features = [
    'blatem', 'blatem-1', 'blactx-m-15', 'sul1', 'sul2', 'tet(a)', 'dfra17' 
    # ... (imagine all 92 are here)
]

detected_genes_from_blast = []

with open("my_blast_results.xml") as result_handle:
    blast_records = NCBIXML.parse(result_handle)
    
    for blast_record in blast_records:
        for alignment in blast_record.alignments:
            for hsp in alignment.hsps:
                # We only want highly confident matches
                if hsp.expect < 0.001: 
                    
                    # Get the database description and make it lowercase
                    db_description = alignment.hit_def.lower()
                    
                    # Check if any of our model's genes are in the description
                    for gene in model_features:
                        if gene in db_description:
                            if gene not in detected_genes_from_blast:
                                detected_genes_from_blast.append(gene)

print("Genes detected by BLAST:")
print(detected_genes_from_blast)
```
